```python
#資料前處例
train_data = pd.read_csv('adult.data', header= None, names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'class'])
print("各項nan記錄加總:")
print("------------------------------------------------")
train_data.apply(lambda x : np.sum(x == " nan"))
train_data = train_data.replace(' ?', np.nan)
train_data.info()
train_data.describe()
```

```python
#各個特徵資料加總紀錄
data_name = np.genfromtxt("adult.names3",delimiter=' ,', dtype=None, encoding='UTF-8')
print("各項特徵值的加總:")
i=0
while i<len(data_name): 
    y=data_name[i]
    y_count = train_data.value_counts(y)
    print(y_count)
    print("------------------------------------------------")
    i=i+1
```

```python
#Nan替換成眾數
train_data['workclass'] = train_data['workclass'].replace(float('nan'),' Private')
train_data['occupation'] = train_data['occupation'].replace(float('nan'),' Prof-specialty')
train_data['native-country'] = train_data['native-country'].replace(float('nan'),' United-States')
```

```python
train_data.isnull().sum(axis=0)
```

```python
#處理名目，先轉換成數字
#One Hot Encoding (Dummies)
data_dum = pd.get_dummies(train_data, dtype=int)
pd.DataFrame(data_dum)
```

```python
#split dataset in features and target variable
feature_cols = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'class']
x = train_data[feature_cols] # Features
y = train_data["class"] # Target variable
# print(x)
# print(y)

# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) # 70% training and 30% test
# print(x_train)
# print(x_test)
# print(y_train)
# print(y_test)

# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(x_test)
```
